# ***15. Top 5 Most Popular Email providers (e.g. gmail.com, yahoo.com,etc...)***

```python
list1 = []
for email in df['Email']:
  list1.append(email.split('@')[1])
```

```python
df['Domain_Name'] = list1
```

```python
df['Domain_Name'].value_counts().head(5)
```
